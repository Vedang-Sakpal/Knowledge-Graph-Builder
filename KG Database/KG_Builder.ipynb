```python
#load the CSV file
import pandas as pd

df = pd.read_csv(
    r'C:\Users\a\Desktop\Internship IITM\Knowledge Graph Builder\HAZOP_006.csv',
    na_values=[''],         # Only empty cells are treated as NaN
    keep_default_na=False   # Do not treat 'None', 'NA', etc. as NaN
)
```

```python
#display the first few rows of the dataframe 
df
```

```python
# Fill NaN values in each column with the previous row's value (forward fill)
df_filled = df.fillna(method='ffill')

# Now you can group or process each row as needed
df_filled
```

# Spliting the data group wise 

Split the data column wise for each row

```python
#calling each row to split the data column wise
for index, row in df_filled.iterrows():
    print(f"Row {index}:")
    for col in df_filled.columns:
        print(f"{col}: {row[col]}")
    print("\n")  # Newline for better readability
```

```python
# Function to escape single quotes for Cypher string literals
def escape_cypher_string(s):
    # Ensure all values are treated as strings before attempting to replace
    return str(s).replace("'", "\\'")

# Initialize an empty list to store Cypher queries
cypher_queries = []

# Iterate through each row of the DataFrame
for index, row in df_filled.iterrows():
    # Escape all string values before using them in Cypher queries
    ref = escape_cypher_string(row['Ref.'])
    parameter = escape_cypher_string(row['Parameter'])
    deviation = escape_cypher_string(row['Deviation'])
    possible_cause = escape_cypher_string(row['Possible Cause'])
    consequence = escape_cypher_string(row['Consequence'])
    safeguard_protection = escape_cypher_string(row['Safeguard/Protection'])
    action_no = escape_cypher_string(row['No'])
    action_description = escape_cypher_string(row['Action'])
    responsible_party = escape_cypher_string(row['On'])

    # Consolidated Cypher query for the entire row's data
    single_line_query = (
        f"MERGE (h:HAZOP_Item {{id: '{ref}'}}) "
        f"MERGE (p:Parameter {{name: '{parameter}'}}) "
        f"MERGE (h)-[:HAS_PARAMETER]->(p) "
        f"MERGE (d:Deviation {{description: '{deviation}'}}) "
        f"MERGE (p)-[:HAS_DEVIATION]->(d) "
        f"MERGE (c:Cause {{description: '{possible_cause}'}}) "
        f"MERGE (d)-[:CAUSED_BY]->(c) "
        f"MERGE (con:Consequence {{description: '{consequence}'}}) "
        f"MERGE (c)-[:LEADS_TO]->(con) "
        f"MERGE (s:Safeguard {{description: '{safeguard_protection}'}}) "
        f"MERGE (con)-[:PROTECTED_BY]->(s) "
        f"MERGE (ai:Action {{id: '{action_no}', description: '{action_description}'}}) "
        f"MERGE (d)-[:HAS_ACTION_ITEM]->(ai);"
    )
    cypher_queries.append(single_line_query)

# Join all queries into a single string, separated by newlines.
# Each line in the final script will correspond to one consolidated MERGE statement for a row of data.
final_cypher_script = "\n".join(cypher_queries)

# Print the first few consolidated lines of the script to verify
print("First 5 consolidated lines of the generated Cypher script:\n")
print("\n".join(final_cypher_script.split('\n')[:5]))

# Save the Cypher script to a file
with open("hazop_knowledge_graph.cypher", "w") as f:
    f.write(final_cypher_script)

print("\nCypher script generated and saved to 'hazop_knowledge_graph.cypher'")
```
