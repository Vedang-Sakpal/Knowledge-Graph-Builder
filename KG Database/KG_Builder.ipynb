# Load the HAZOP report 

The file can be xlxs or CSV derive the different types of rows and split them into particular groups.


### The Groups are as :
- Equipment 
- Parameter 
- Devitation 
- Causes 
- Consequences 
- Safeguards 

Further these can be used to create the Knowledge Graphs 

```python
#load the CSV file
import pandas as pd

df = pd.read_csv(
    r'C:\Users\a\Desktop\Internship IITM\Knowledge Graph Builder\HAZOP_006.csv',
    na_values=[''],         # Only empty cells are treated as NaN
    keep_default_na=False   # Do not treat 'None', 'NA', etc. as NaN
)
```

```python
#display the first few rows of the dataframe 
df.head()
```

```python
# Fill NaN values in each column with the previous row's value (forward fill)
df_filled = df.fillna(method='ffill')

# Now you can group or process each row as needed
df_filled.head()
```

# Spliting the data group wise 

Split the data column wise for each row

```python
#calling each row to split the data column wise
for index, row in df_filled.iterrows():
    print(f"Row {index}:")
    for col in df_filled.columns:
        print(f"{col}: {row[col]}")
    print("\n")  # Newline for better readability
```

```python
# Function to escape single quotes and other special characters for Cypher string literals
def escape_cypher_string(s):
    """Escapes special characters in a string for use in Cypher."""
    # Ensure all values are treated as strings before escaping
    s = str(s)
    return s.replace("\\", "\\\\").replace("'", "\\'").replace('"', '\\"')

# Initialize an empty list to store Cypher queries
cypher_queries = []

# Iterate through each row of the DataFrame
for index, row in df_filled.iterrows():
    # Escape all string values to prevent Cypher injection and errors
    equipment = escape_cypher_string(row['equipment'])
    parameter = escape_cypher_string(row['Parameter'])
    deviation = escape_cypher_string(row['Deviation'])
    type_of_cause = escape_cypher_string(row['Type of Cause'])
    possible_cause = escape_cypher_string(row['Possible Cause'])
    type_of_consequence = escape_cypher_string(row['Type of Consequence'])
    consequence = escape_cypher_string(row['Consequence'])
    type_of_safeguard = escape_cypher_string(row['Type of Safeguard'])
    safeguard = escape_cypher_string(row['Safeguard/Protection'])
    action_no = escape_cypher_string(row['No'])
    action_desc = escape_cypher_string(row['Action'])
    responsible_party = escape_cypher_string(row['On'])

    # Build the Cypher query for the current row with a corrected graph model.
    # Each MERGE statement is on its own line for clarity.
    query = (
        # 1. MERGE all nodes first to ensure they exist before creating relationships
        f"MERGE (e:Equipment {{name: '{equipment}'}}) "
        f"MERGE (p:Parameter {{name: '{parameter}'}}) "
        f"MERGE (d:Deviation {{description: '{deviation}'}}) "
        f"MERGE (c:Cause {{description: '{possible_cause}'}}) "
        f"MERGE (tc:TypeOfCause {{name: '{type_of_cause}'}}) "
        f"MERGE (con:Consequence {{description: '{consequence}'}}) "
        f"MERGE (tcon:TypeOfConsequence {{name: '{type_of_consequence}'}}) "
        f"MERGE (s:Safeguard {{description: '{safeguard}'}}) "
        f"MERGE (ts:TypeOfSafeguard {{name: '{type_of_safeguard}'}}) "
        # f"MERGE (a:Action {{id: '{action_no}', description: '{action_desc}', responsibleParty: '{responsible_party}'}}) "

        # 2. MERGE the relationships between the nodes to build the logical graph
        f"MERGE (e)-[:HAS_PARAMETER]->(p) "
        f"MERGE (p)-[:HAS_DEVIATION]->(d) "
        # Deviation is caused by the specific Cause
        f"MERGE (d)-[:CAUSED_BY]->(c) "
        # The specific Cause is categorized by a TypeOfCause
        f"MERGE (c)-[:IS_A]->(tc) "
        # The specific Cause leads to a specific Consequence
        f"MERGE (c)-[:LEADS_TO]->(con) "
        # The specific Consequence is categorized by a TypeOfConsequence
        f"MERGE (con)-[:IS_A]->(tcon) "
        # The specific Consequence is protected by a specific Safeguard
        f"MERGE (con)-[:PROTECTED_BY]->(s) "
        # The specific Safeguard is categorized by a TypeOfSafeguard
        f"MERGE (s)-[:IS_A]->(ts) "
        # The Equipment has a specific Safeguard
        f"MERGE (e)-[:HAS_SAFEGUARD]->(s); "
        # The Deviation recommends a specific Action
        # f"MERGE (d)-[:RECOMMENDS_ACTION]->(a);"
    )
    cypher_queries.append(query)

# Join all queries into a single script string
final_cypher_script = "\n".join(cypher_queries)

# Print the first query to verify its structure
print("First generated Cypher query:\n")
print(final_cypher_script.split('\n')[0])

# Save the corrected Cypher script to a new file
with open("hazop_knowledge_graph_corrected.cypher", "w") as f:
    f.write(final_cypher_script)

print("\n✅ Corrected Cypher script generated and saved to 'hazop_knowledge_graph_corrected.cypher'")
```
